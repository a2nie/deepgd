```python
def exp_smooth(array, tau):
    def exp_avg(array, tau):
        w = np.flip(np.power(tau, np.arange(len(array))))
        return np.dot(w, array) / w.sum()
    return np.array([exp_avg(array[:i+1], tau) for i in range(len(array))])
```

```python
def soft_adapt(losses, tau=0.95, beta=1, eps=1e-8):
    smoothed = np.stack(list(map(lambda l: exp_smooth(l, tau), losses)), axis=1)
    initial_rate = np.ones([1, smoothed.shape[1]])
    roc = np.concatenate([initial_rate, -np.diff(smoothed, axis=0)], axis=0)
    normalized_roc = roc / (np.abs(roc).sum(axis=1)[:, None] + eps)
    exp_beta_roc = np.exp(beta * normalized_roc)
    loss_weighted = smoothed * exp_beta_roc
    return loss_weighted / loss_weighted.sum(axis=1)[:, None]
```
