```python
from bentley_ottmann.planar import segments_intersections as intersect
from lib.utils import *
from lib.modules import *
```

```python
def find_intersect(segments,accurate = True):
    return intersect(segments,accurate = accurate)

def graph_vis(G, node_pos, file_name=None):
    i = 0
    for n, p in node_pos:
        node = 'n' +str(i)
        G.nodes[node]['pos'] = (n,p)
        i += 1
    pos = nx.get_node_attributes(G,'pos')
    plt.figure()
    nx.draw(G, pos)
    if file_name is not None:
        plt.savefig(file_name)
```
