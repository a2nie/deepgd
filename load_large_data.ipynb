```python
def generate_apsp(G):
    n = G.number_of_nodes()
    apsp_gen = nx.all_pairs_shortest_path_length(G)
    return np.array([np.fromiter(j.values(), int)[np.argsort(np.fromiter(j.keys(), int))] for i, j in tqdm(apsp_gen, total=n)])
```
