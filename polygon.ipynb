```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_polygon(n, radius=1):
    return [[(radius * np.cos(2 * np.pi * i / n),
                      radius * np.sin(2 * np.pi * i / n)) for i in range(n)]]
```

```python
nodes = generate_polygon(100)
```
